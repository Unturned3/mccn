```python
# facenet_pytorch_c: avoid confusion with system default facenet_pytorch
#from facenet_pytorch_c import MTCNN

from tqdm import tqdm
import numpy as np
import os

# pytorch
import torch
from torch import nn

# data handling
from torch.utils.data import DataLoader

# torchvision libs
import torchvision
from torchvision import datasets
from torchvision import transforms
import PIL
from PIL import Image

import utils

import matplotlib.pyplot as plt
import matplotlib.patches as patches

#device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device = 'cpu'

# data loading parameters
workers = 4

x_t, b_prob_t, b_box_t, f_prob_t, f_box_t, x_v, b_prob_v, b_box_v, f_prob_v, f_box_v = utils.get_images(
    '/home/ubuntu/db_proc/db/images',
    '/home/ubuntu/db_proc/db/annotations',
    valid_percent=100,
    resize_shape=(400,400) # keep original size
)

class PNet(nn.Module):

    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(3, 10, kernel_size=3)
        self.prelu1 = nn.PReLU(10)
        self.pool1 = nn.MaxPool2d(2, 2, ceil_mode=True)
        self.conv2 = nn.Conv2d(10, 16, kernel_size=3)
        self.prelu2 = nn.PReLU(16)
        self.conv3 = nn.Conv2d(16, 32, kernel_size=3)
        self.prelu3 = nn.PReLU(32)
        
        self.conv4_1 = nn.Conv2d(32, 2, kernel_size=1) # body prob
        self.softmax4_1 = nn.Softmax(dim=1)
        
        self.conv4_2 = nn.Conv2d(32, 4, kernel_size=1) # body bbox
        
        self.conv4_3 = nn.Conv2d(32, 2, kernel_size=1) # face prob
        self.softmax4_3 = nn.Softmax(dim=1)
        
        self.conv4_4 = nn.Conv2d(32, 4, kernel_size=1) # face bbox
        
        state_dict = torch.load('pnet.pt')
        #state_dict = torch.load('cd_pnet.pt')
        self.load_state_dict(state_dict)
        
        self.training = True

    def forward(self, x):
        x = self.conv1(x)
        x = self.prelu1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.prelu2(x)
        x = self.conv3(x)
        x = self.prelu3(x)
        
        a = self.conv4_1(x) # body prob
        a = self.softmax4_1(a)
        
        b = self.conv4_2(x) # body bbox
        
        c = self.conv4_3(x) # face prob
        c = self.softmax4_3(c)
        
        d = self.conv4_4(x) # face bbox
        
        return a, b, c, d

class BNet(nn.Module):

    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(3, 32, kernel_size=3)
        self.prelu1 = nn.PReLU(32)
        self.pool1 = nn.MaxPool2d(3, 2, ceil_mode=True)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.prelu2 = nn.PReLU(64)
        self.pool2 = nn.MaxPool2d(3, 2, ceil_mode=True)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3)
        self.prelu3 = nn.PReLU(64)
        self.pool3 = nn.MaxPool2d(2, 2, ceil_mode=True)
        self.conv4 = nn.Conv2d(64, 128, kernel_size=2)
        self.prelu4 = nn.PReLU(128)
        self.dense5 = nn.Linear(1152, 256)
        self.prelu5 = nn.PReLU(256)
        
        self.dense6_1 = nn.Linear(256, 2)    # body prob
        self.softmax6_1 = nn.Softmax(dim=1)
        
        self.dense6_2 = nn.Linear(256, 4)    # body bbox
        
        state_dict = torch.load('bnet.pt')
        self.load_state_dict(state_dict)

        self.training = True

    def forward(self, x):
        x = self.conv1(x)
        x = self.prelu1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.prelu2(x)
        x = self.pool2(x)
        x = self.conv3(x)
        x = self.prelu3(x)
        x = self.pool3(x)
        x = self.conv4(x)
        x = self.prelu4(x)
        x = self.dense5(x.view(x.shape[0], -1))
        x = self.prelu5(x)
        a = self.dense6_1(x)
        a = self.softmax6_1(a)
        b = self.dense6_2(x)
        return a, b

class FNet(nn.Module):
    """MTCNN ONet"""

    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(3, 32, kernel_size=3)
        self.prelu1 = nn.PReLU(32)
        self.pool1 = nn.MaxPool2d(3, 2, ceil_mode=True)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.prelu2 = nn.PReLU(64)
        self.pool2 = nn.MaxPool2d(3, 2, ceil_mode=True)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3)
        self.prelu3 = nn.PReLU(64)
        self.pool3 = nn.MaxPool2d(2, 2, ceil_mode=True)
        self.conv4 = nn.Conv2d(64, 128, kernel_size=2)
        self.prelu4 = nn.PReLU(128)
        self.dense5 = nn.Linear(1152, 256)
        self.prelu5 = nn.PReLU(256)
        self.dense6_1 = nn.Linear(256, 2)
        self.softmax6_1 = nn.Softmax(dim=1) # face prob
        self.dense6_2 = nn.Linear(256, 4)   # face bbox
        self.dense6_3 = nn.Linear(256, 10)  # landmarks

        self.training = False
        
        state_dict = torch.load('onet.pt')
        self.load_state_dict(state_dict)
        
        #del self.dense6_3 # remove landmark task

    def forward(self, x):
        x = self.conv1(x)
        x = self.prelu1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.prelu2(x)
        x = self.pool2(x)
        x = self.conv3(x)
        x = self.prelu3(x)
        x = self.pool3(x)
        x = self.conv4(x)
        x = self.prelu4(x)
        x = x.permute(0, 3, 2, 1).contiguous()
        x = self.dense5(x.view(x.shape[0], -1))
        x = self.prelu5(x)
        a = self.dense6_1(x)
        a = self.softmax6_1(a) # face prob
        b = self.dense6_2(x)   # face bbox
        return a, b

class ANet(nn.Module):
    def __init__(self, class_num):
        super().__init__()

        self.conv1 = nn.Conv2d(3, 32, kernel_size=3)
        self.prelu1 = nn.PReLU(32)
        self.pool1 = nn.MaxPool2d(3, 2, ceil_mode=True)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.prelu2 = nn.PReLU(64)
        self.pool2 = nn.MaxPool2d(3, 2, ceil_mode=True)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3)
        self.prelu3 = nn.PReLU(64)
        self.pool3 = nn.MaxPool2d(2, 2, ceil_mode=True)
        self.conv4 = nn.Conv2d(64, 128, kernel_size=2)
        self.prelu4 = nn.PReLU(128)
        self.dense5 = nn.Linear(1152, 256)
        self.prelu5 = nn.PReLU(256)
        self.dense6_1 = nn.Linear(256, class_num) # age
        
        state_dict = torch.load('8_s.pt')['state_dict']
        self.load_state_dict(state_dict)

        self.training = True

    def forward(self, x):
        x = self.conv1(x)
        x = self.prelu1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.prelu2(x)
        x = self.pool2(x)
        x = self.conv3(x)
        x = self.prelu3(x)
        x = self.pool3(x)
        x = self.conv4(x)
        x = self.prelu4(x)
        x = self.dense5(x.view(x.shape[0], -1))
        x = self.prelu5(x)
        a = self.dense6_1(x)
        return a


pnet, bnet, fnet, anet = PNet(), BNet(), FNet(), ANet(8)
pnet.eval()
bnet.eval()
fnet.eval()
anet.eval()
pnet.to(device);
bnet.to(device);
fnet.to(device);
anet.to(device);
```

```python

# paramters
idx = 19
factor = 0.709
#factor = 0.502681
thrs = [0.7, 0.9]
plt.imshow(x_v[idx])
plt.show()
```

```python
display = False
t0 = time.time()

########################### generate image pyramid ############################

o_img = x_v[idx]
o_width, o_height = o_img.size
t_scale = 1

ipymd, spymd = [], []

while min(o_width, o_height) >= 12:
    
    ipymd.append(o_img)
    spymd.append(t_scale)
    
    t_scale *= factor
    o_width *= factor
    o_height *= factor
    o_img = o_img.resize((int(round(o_width)), int(round(o_height))), Image.LANCZOS)

o_img = ipymd[0]
o_width, o_height = o_img.size





######################## run PNet over input image ############################




lastn = -6

with torch.no_grad():
    bp, bb, fp, fb = [], [], [], []

    for img, scale in zip(ipymd[lastn:], spymd[lastn:]):

        # unsqueeze: add 1 extra "batch" dimension to image
        img = transforms.ToTensor()(img).unsqueeze(0).to(device)
        obp, obb, ofp, ofb  = pnet(img)

        # tensor.shape = (..., height, width)

        obp, obb, ofp, ofb = obp.squeeze(), obb.squeeze(), ofp.squeeze(), ofb.squeeze()
        
        obp, ofp = obp[0, :], ofp[0, :]  # remove the dimension indicating probability of no body/face

        bp_mask, fp_mask = obp >= thrs[0], ofp >= thrs[0]
        bpm_inds, fpm_inds = bp_mask.nonzero(), fp_mask.nonzero()
        obp, obb, ofp, ofb = obp[bp_mask], obb[:, bp_mask], ofp[fp_mask], ofb[:, fp_mask]

        obb, ofb = obb.permute(1, 0), ofb.permute(1, 0)
        
        obb[:, 0] = (obb[:, 0] + bpm_inds[:, 1] * 2) / scale
        obb[:, 1] = (obb[:, 1] + bpm_inds[:, 0] * 2) / scale
        obb[:, 2] = (obb[:, 2] + bpm_inds[:, 1] * 2) / scale
        obb[:, 3] = (obb[:, 3] + bpm_inds[:, 0] * 2) / scale
 
        ofb[:, 0] = (ofb[:, 0] + fpm_inds[:, 1] * 2) / scale
        ofb[:, 1] = (ofb[:, 1] + fpm_inds[:, 0] * 2) / scale
        ofb[:, 2] = (ofb[:, 2] + fpm_inds[:, 1] * 2) / scale
        ofb[:, 3] = (ofb[:, 3] + fpm_inds[:, 0] * 2) / scale

        bp.append(obp)
        bb.append(obb)
        fp.append(ofp)
        fb.append(ofb)

    t_bp = torch.Tensor(len(bp))
    t_bb = torch.Tensor(len(bb), 4)
    t_fp = torch.Tensor(len(fp))
    t_fb = torch.Tensor(len(fb), 4)

    torch.cat(bp, out=t_bp)
    torch.cat(bb, out=t_bb)
    torch.cat(fp, out=t_fp)
    torch.cat(fb, out=t_fb)




############################### process PNet output #################################




bids = torchvision.ops.nms(t_bb, t_bp, 0.5)
fids = torchvision.ops.nms(t_fb, t_fp, 0.5)

bps = t_bp[bids].numpy()
bbs = t_bb[bids].numpy()

fps = t_fp[fids].numpy()
fbs = t_fb[fids].numpy()

# convert bbox output of PNet to squares, and expand them in each direction
epad = 25
po_b, co_b, po_f, co_f = [], [], [], []

for i in range(0, len(bbs)):
    cw = bbs[i, 2] - bbs[i, 0]
    ch = bbs[i, 3] - bbs[i, 1]
    diff = int(round(abs(cw-ch)/2))
    if True:
        if cw > ch:
            bbs[i, 1] -= diff
            bbs[i, 3] += diff
        else:
            bbs[i, 0] -= diff
            bbs[i, 2] += diff
    crp = o_img.crop((bbs[i,0]-epad, bbs[i,1]-epad, bbs[i,2]+epad, bbs[i,3]+epad))
    po_b.append(crp)
    co_b.append(np.array([bbs[i,0]-epad, bbs[i,1]-epad, bbs[i,2]+epad, bbs[i,3]+epad], dtype='float32'))

# convert fbox output of PNet to squares, and expand them in each direction
for i in range(0, len(fbs)):
    cw = fbs[i, 2] - fbs[i, 0]
    ch = fbs[i, 3] - fbs[i, 1]
    diff = int(round(abs(cw-ch)/2))
    if True:
        if cw > ch:
            fbs[i, 1] -= diff
            fbs[i, 3] += diff
        else:
            fbs[i, 0] -= diff
            fbs[i, 2] += diff
    crp = o_img.crop((fbs[i,0]-epad, fbs[i,1]-epad, fbs[i,2]+epad, fbs[i,3]+epad))
    po_f.append(crp)
    co_f.append(np.array([fbs[i,0]-epad, fbs[i,1]-epad, fbs[i,2]+epad, fbs[i,3]+epad], dtype='float32'))

po_b = [k.resize((48,48), Image.LANCZOS) for k in po_b]
po_f = [k.resize((48,48), Image.LANCZOS) for k in po_f]

found_body, found_face = True, True

if len(po_b) == 0:
    found_body = False
if len(po_f) == 0:
    found_face = False



################################### second stage ###################################



### NOTE
# nms is ineffective at this stage. coords need to be converted back to global.

if found_body:
    # unsqueeze: add 1 extra "batch" dimension to image
    imgs = [transforms.ToTensor()(k).to(device) for k in po_b]
    imgs = torch.stack(imgs) # stack all possible body images into 1 tensor (batch process)
    #co_b = [transforms.ToTensor()(np.expand_dims(k, 0)).to(device) for k in co_b]
    #co_b = torch.stack(co_b)

    obp, obb = bnet(imgs) # ignore output about face (since all these are potential "body" regions)

    #obp, obb = obp.squeeze(), obb.squeeze()
    obp = obp[:, 0] # remove dimension for "negative" probability (and keep batch dimension)

    bp_mask = obp >= 0.7
    #bpm_inds = bp_mask.nonzero()
    obp, obb, imgs = obp[bp_mask], obb[bp_mask, :], imgs[bp_mask, :]

    co_b = np.array(co_b)
    co_b = co_b[bp_mask.detach().numpy()]
    #co_b = co_b.tolist()
    #co_b = [co_b[k] for k in bp_mask.detach().numpy().tolist()]

    # wrong place to do NMS...
    """
    keep_inds = torchvision.ops.nms(obb, obp, 0.3)
    obb, obp = obb[keep_inds].detach().numpy(), obp[keep_inds].detach().numpy()

    imgs = imgs[keep_inds]
    """
    obb, obp = obb.detach().numpy(), obp.detach().numpy()
    imgs = [torchvision.transforms.ToPILImage()(k) for k in imgs]

    final_body, final_bprob = [], []

    #print("detected body numbers: {}".format(len(imgs)))
    for k, b, p, ab in zip(imgs, obb, obp, co_b):
        """
        fig,ax = plt.subplots(1)
        ax.imshow(k)
        rect = patches.Rectangle((b[0],b[1]),b[2]-b[0],b[3]-b[1],linewidth=1,edgecolor='r',facecolor='none')
        ax.add_patch(rect)
        plt.show()
        print("bbox: {}".format(b.round(0)))
        """

        # turn bbox into square

        cw = b[2] - b[0]
        ch = b[3] - b[1]
        diff = (abs(cw-ch)/2)
        if True:
            if cw > ch:
                b[1] -= diff
                b[3] += diff
            else:
                b[0] -= diff
                b[2] += diff

        aw, ah = ab[2]-ab[0], ab[3]-ab[1]

        b[0] = b[0] / 48 * aw + ab[0]
        b[1] = b[1] / 48 * ah + ab[1]
        b[2] = b[2] / 48 * aw + ab[0]
        b[3] = b[3] / 48 * ah + ab[1]

        final_body.append(b)
        final_bprob.append(p)

    #final_body = [transforms.ToTensor()(k).to(device) for k in final_body]
    final_body = transforms.ToTensor()(np.array([final_body])).permute(2, 1, 0).squeeze(1).to(device)
    #final_body = torch.stack(final_body)
    #final_bprob = [transforms.ToTensor()(k).to(device) for k in final_bprob]
    final_bprob = transforms.ToTensor()(np.array([final_bprob])).permute(2, 1, 0).squeeze(1).to(device)
    #final_bprob = torch.stack(final_bprob)

    keep_inds = torchvision.ops.nms(final_body, final_bprob, 0.1)
    final_body, final_bprob = final_body[keep_inds].detach().numpy(), final_bprob[keep_inds].detach().numpy()
    #final_body, final_bprob = final_body.detach().numpy(), final_bprob.detach().numpy()

    #imgs = imgs[keep_inds]
    if display:
        print("# of body after NMS: {}".format(len(final_body)))
        for b, p in zip(final_body, final_bprob):
            plt.imshow(o_img.crop(tuple(b.squeeze())))
            plt.show()
            print(p)


if found_face:
    # caution: FNet doesn't output raw coordinates for bbox. They are in range (-1, 1)

    # unsqueeze: add 1 extra "batch" dimension to image
    imgs = [transforms.ToTensor()(k).to(device) for k in po_f]
    imgs = torch.stack(imgs) # stack all possible body images into 1 tensor (batch process)

    ofp, ofb = fnet(imgs) # ignore output about face (since all these are potential "body" regions)
    #obp, obb = obp.squeeze(), obb.squeeze()
    ofp = ofp[:, 1] # remove dimension for "negative" probability (and keep batch dimension)

    fp_mask = ofp >= 0.3
    #fpm_inds = fp_mask.nonzero()
    ofp, ofb, imgs = ofp[fp_mask], ofb[fp_mask, :], imgs[fp_mask, :]

    co_f = np.array(co_f)
    co_f = co_f[fp_mask.detach().numpy()]
    #co_f = co_f.tolist()

    ofb[:, 0] = ofb[:, 0] * 48
    ofb[:, 1] = ofb[:, 1] * 48
    ofb[:, 2] = 48 + 48 * ofb[:, 2]
    ofb[:, 3] = 48 + 48 * ofb[:, 3]

    # wrong place to do NMS, skip for now...
    """
    keep_inds = torchvision.ops.nms(ofb, ofp, 0.3)
    ofb, ofp = ofb[keep_inds].detach().numpy(), ofp[keep_inds].detach().numpy()

    imgs = imgs[keep_inds]
    """

    ofb, ofp = ofb.detach().numpy(), ofp.detach().numpy()

    imgs = [torchvision.transforms.ToPILImage()(k) for k in imgs]
    final_face, final_fprob = [], []
    aimgs = []

    #print("detected face numbers: {}".format(len(imgs)))
    for k, b, p, ab in zip(imgs, ofb, ofp, co_f):
        """
        fig,ax = plt.subplots(1)
        ax.imshow(k)
        rect = patches.Rectangle((b[0],b[1]),b[2]-b[0],b[3]-b[1],linewidth=1,edgecolor='r',facecolor='none')
        ax.add_patch(rect)
        plt.show()
        print("prob: {:.2f}; bbox: {}".format(p, b.round(0)))
        """
        
        # prepare images for age classification

        # turn bbox into square

        cw = b[2] - b[0]
        ch = b[3] - b[1]
        diff = (abs(cw-ch)/2)
        if True:
            if cw > ch:
                b[1] -= diff
                b[3] += diff
            else:
                b[0] -= diff
                b[2] += diff

        aw, ah = ab[2]-ab[0], ab[3]-ab[1]
        
        b[0] = b[0] / 48 * aw + ab[0]
        b[1] = b[1] / 48 * ah + ab[1]
        b[2] = b[2] / 48 * aw + ab[0]
        b[3] = b[3] / 48 * ah + ab[1]
        
        final_face.append(b)
        final_fprob.append(p)
        #aimgs.append(o_img.crop((b[0], b[1], b[2], b[3])))
    
    #final_body = [transforms.ToTensor()(k).to(device) for k in final_body]
    final_face = transforms.ToTensor()(np.array([final_face])).permute(2, 1, 0).squeeze(1).to(device)
    #final_body = torch.stack(final_body)
    #final_bprob = [transforms.ToTensor()(k).to(device) for k in final_bprob]
    final_fprob = transforms.ToTensor()(np.array([final_fprob])).permute(2, 1, 0).squeeze(1).to(device)
    #final_bprob = torch.stack(final_bprob)

    keep_inds = torchvision.ops.nms(final_face, final_fprob, 0.1)
    final_face, final_fprob = final_face[keep_inds].detach().numpy(), final_fprob[keep_inds].detach().numpy()
    #final_body, final_bprob = final_body.detach().numpy(), final_bprob.detach().numpy()

    #imgs = imgs[keep_inds]
    if display:
        print("# of face after NMS: {}".format(len(final_face)))
        for b, p in zip(final_face, final_fprob):
            plt.imshow(o_img.crop(tuple(b.squeeze())))
            plt.show()
            print(p)

final_age = []

for b in final_face:
    i = o_img.crop((b[0], b[1], b[2], b[3]))
    i = i.resize((48, 48), Image.LANCZOS)
    age = anet(transforms.ToTensor()(i).unsqueeze(0).to(device))
    final_age.append(torch.max(age.data, 1)[1].item())
    if display:
        plt.imshow(i)
        plt.show()
        print(torch.max(age.data, 1)[1].item())

    
t1 = time.time()
total = t1-t0
print("time taken: {:.2f} seconds".format(total))
```
